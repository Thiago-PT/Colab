```python
# Ejercicio No. 4: Leer un texto y determinar cuántas veces dicho texto contiene cada una de las vocales en minúscula y mayúscula.

Texto = input("Digite el texto: ")

minus = {'a' : 0, 'e' : 0, 'i' : 0, 'o' : 0, 'u' : 0 }
mayus = {'A' : 0, 'E' : 0, 'I' : 0, 'O' : 0, 'U' : 0 }

for i in Texto:
  if i in minus:
    minus[i] += 1
  if i in mayus:
    mayus[i] +=1

print("Vocales en minúscula: ")
print(minus)

print("Vocales en mayuscúla: ")
print(mayus)
```

```python
# Ejercicio No. 6: Leer 20 números enteros y averiguar cuántos son pares y cuántos son impares.

pares = 0
impares = 0

for i in range(20):
    número = int(input("Introduzca un número entero: "))
    if número % 2 == 0:
        pares += 1
    else:
        impares += 1

print("Cantidad de números pares:", pares)
print("Cantidad de números impares:", impares)
```

```python
# Ejercicio No. 7: Programa que averigüe e imprima cuántos múltiplos de 7, y cuántos múltiplos de 9, hay en los números comprendidos entre 1000 y 5000.

multiplos_7 = 0
multiplos_9 = 0

for i in range(1000, 5001):
    if i % 7 == 0:
        multiplos_7 += 1
    if i % 9 == 0:
        multiplos_9 += 1

print("Cantidad de múltiplos de 7:", multiplos_7)
print("Cantidad de múltiplos de 9:", multiplos_9)
```

```python
# Ejercicio No. 9: Programa para calcular el factorial de un número.

print ("Calcula el factorial")
numero = int(input("Introduzca el número: "))

factorial = 1

i = 1
while (i <= numero):
    factorial = factorial * i
    i = i + 1

print ("El factorial de " + str(numero) + " es " + str(factorial))
```
